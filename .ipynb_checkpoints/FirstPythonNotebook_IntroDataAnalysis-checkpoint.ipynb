```python
x = [1, 3, 5]
y = [2, 4, 6]
plt.plot(x,y)
plt.title("hooray lines")
plt.xlabel("bottom")
plt.ylabel("side")
plt.show
```

```python
x = [1, 3, 5]
y = [2, 4, 6]
z = [11, 7, 3]
plt.plot(x,y)
plt.plot(x,z)
plt.title("hooray lines")
plt.xlabel("bottom")
plt.ylabel("side or edge")
plt.show
```

```python
x = [1, 3, 5]
y = [2, 4, 6]
z = [11, 7, 3]
plt.plot(x,y)
plt.plot(x,z)
plt.title("hooray lines")
plt.xlabel("bottom")
plt.ylabel("side or edge")
plt.legend(["This is blue", "This is orange"])
plt.show;
```

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
us = data[data.country == 'United States']
```

```python
us
```

```python
data.country == 'United States'
```

```python
data[data.country == 'United States']
```

```python
china = data[data.country == 'China']
```

```python
china
```

```python
plt.plot(us.year, us.population)
plt.show();
```

```python
plt.plot(us.year, us.population / 10**6)
plt.show();
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.show();
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.show();
```
